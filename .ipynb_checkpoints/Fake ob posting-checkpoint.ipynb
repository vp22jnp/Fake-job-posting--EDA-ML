```python
x= df.iloc[:,:-1]
y=df['fraudulent']
```

```python
from sklearn.model_selection import train_test_split
x_train , y_train , x_test , y_test = train_test_split (x,y , test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.preprocessing import LabelEncoder
paymentMethod_le=LabelEncoder() 
df["title"]=paymentMethod_le.fit_transform(df["title"])
```

```python
df["description"]=paymentMethod_le.fit_transform(df["description"])
df["requirements"]=paymentMethod_le.fit_transform(df["requirements"])
df["company_profile"]=paymentMethod_le.fit_transform(df["company_profile"])
df["location"]=paymentMethod_le.fit_transform(df["location"])
df["salary_range"]=paymentMethod_le.fit_transform(df["salary_range"])
df["employment_type"]=paymentMethod_le.fit_transform(df["employment_type"])
df["industry"]=paymentMethod_le.fit_transform(df["industry"])
df["benefits"]=paymentMethod_le.fit_transform(df["benefits"])


```
